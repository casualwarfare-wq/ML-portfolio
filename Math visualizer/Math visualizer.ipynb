```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
x = np.arange(0, 10)
xl = np.arange(0.1, 10, 0.5)
fig, ax = plt.subplots(2,2,figsize=(16, 8))
ax[0][0].plot(x, np.e**x, label='y = e^x', marker = 'x', mfc = 'purple', mec = 'blue', color = 'red')
ax[0][1].plot(xl, np.log(xl), label='y = log(x)', marker = 'o', mfc = 'yellow', mec = 'red', color = 'orange')
ax[1][0].plot(x, x, label='y = x', marker = '.', mfc = 'purple', mec = 'cyan', color = 'blue')
ax[0][0].set_xlabel('powers')
ax[0][0].set_ylabel('values')
ax[0][1].set_xlabel('values')
ax[0][1].set_ylabel('log(values)')
ax[1][0].set_xlabel('x')
ax[1][0].set_ylabel('y')
ax[1][1].plot(x, np.cbrt(x), label='y = cbrt(x)', marker = '*', mfc = 'green', mec = '#86d021', color = 'purple')
ax[1][1].set_xlabel('x')
ax[1][1].set_ylabel('cbrt(x)')
ax[0][0].grid(axis = 'both')
ax[0][1].grid(axis = 'both')
ax[1][0].grid(axis = 'both')
ax[1][1].grid(axis = 'both')
fig.legend()
fig.suptitle('Exponential and Logarithmic Functions')
print('Exponential and Logarithmic Functions explain loss functions in machine learning.  The logarithmic function is used in loss functions like cross-entropy loss.')
```

```python
x = np.linspace(0, 5, 50)
der = 3*(x**2)
fig, ax = plt.subplots(2,1,figsize=(16, 8))
fig.suptitle('Function and its Derivative')
ax[0].plot(x, x**3, label='y = x^3', marker = 'o', mfc = 'blue', mec = 'purple', color = 'orange')
ax[0].plot(x, der, label = 'dx/dy = 3x^2',color = 'red', linestyle = '-.')
ax[0].set_xlabel('powers')
ax[0].set_ylabel('values')
ax[0].legend()
ax[0].grid(True)
ax[0].set_title('Function and secant lines')
dx = 4
tan = np.log(3)*(3**dx)
tan = 3**dx + tan*(x-dx)
ax[1].plot(x, 3**x, label='y = 3^x', marker = 'x', mfc = 'red', mec = 'orange', color = 'yellow')
ax[1].plot(x, tan, label='Tangent Line', color = 'blue', linestyle = '--')
ax[1].set_xlabel('powers')
ax[1].set_ylabel('values')
ax[1].legend()
ax[1].grid(True)
ax[1].set_title('Exponential Function and Tangent Lines')
fig.tight_layout()
print('Derivatives are used in model tuning, optimization and the chain rule is used in backpropagation in machine learning.')
```

```python
x = np.random.normal(loc=67, scale=420, size=1000)
plt.hist(x, bins = 20, color = 'orange', edgecolor = 'red', density = True)
xk = np.linspace(-1000, 1000, 10000)
plt.plot(xk, stats.gaussian_kde(x)(xk), color = 'blue', linewidth = 2)
print('Probability distributions used used to model uncertainty in machine learning.  The normal distribution is used in Bayesian inference and Gaussian processes.')
```

```python
x = np.linspace(-10, 10, 1000)
mu, std = stats.norm.fit(x)
plt.plot(x, stats.norm.pdf(x, mu, std), color = 'blue')
print('The normal distribution is used to set weights in neural networks and to model uncertainty in machine learning as well as modeling errors and noise.  The normal distribution is used in Bayesian inference and Gaussian processes.')
```
